```python
import numpy as np
import matplotlib.pyplot as plt
```

# Unit Discrete step signal is given by

## $$u[n-a]=$$
\begin{cases}
1 & n \ge a \\
0 & n < a
\end{cases}

```python
#plot Discrete u[n-n0]
def Unit_step(n,n0):
    unit = []
    for i in n:
        if i<n0:
            unit.append(0)
        else:
            unit.append(1)
    return unit
LL = -10
UL = 20
n = np.arange(LL,UL,1)  #n[-10,-9,-8,...,0,1,2,3,...,9] 
shift_time = 4        #u[n-4]
unit=Unit_step(n,shift_time)  

plt.stem(n,unit)
plt.xlabel('n')
plt.xticks(np.arange(LL, UL, 1))
plt.yticks([0, 1])
plt.ylabel('u[n]')
plt.title('Unit step u[n-a]')
plt.show()
```

# Discrete Unit impulse signal is given by

$$\delta(n-a) = $$
\begin{cases}
1 & n = a \\
0 & n \neq a
\end{cases}

```python
def Impulse_unit(n,a):
    impulse = []
    for i in n:
        if i==a:
            impulse.append(1)
        else:
            impulse.append(0)
            
    return impulse

LL = -10
UL = 20
n = np.arange(LL,UL,1)  #n[-10,-9,-8,...,0,1,2,3,...,9] 
shift_time = 4        #u[n-4]
impulse=Impulse_unit(n,shift_time)  

plt.stem(n,impulse)
plt.xlabel('n')
#plt.xticks(np.arange(LL, UL, 1))
plt.yticks([0, 1])
plt.ylabel('$\delta[n]$')
plt.title('Discrete Unit impulse signal $\delta[n-a]$')
plt.show()
```

# Unit Ramp Signal
$$r[n]$$ 
\begin{cases}
n & n >= a \\
0 & n < a
\end{cases}

```python
def Ramp_signal(n,a):
    ramp = []
    for i in n:
        if i<a:
            ramp.append(0)
        else:
            ramp.append(i)
            
    return ramp

LL = -10
UL = 20
n = np.arange(LL,UL,1)  #n[-10,-9,-8,...,0,1,2,3,...,9] 
shift_time = 4        #n0=4

ramp=Ramp_signal(n,shift_time)  #r[n-4]
plt.stem(n,ramp)
plt.xlabel('n')
plt.xticks(np.arange(LL, UL, 1))
plt.yticks([0, n[-1]])
plt.ylabel('r[n]')
plt.title('Unit Ramp Signal r[n-a]')
plt.show()
```

# Continus Unit Ramp Signal

$$r(t)$$ 
\begin{cases}
t & t >= a \\
0 & t < a
\end{cases}

```python

def Ramp_signal_continus(n,a):
    ramp = []
    for i in n:
        if i<a:
            ramp.append(0)
        else:
            ramp.append(i)
            
    return ramp

LL = -10
UL = 200
n = np.arange(LL,UL,.1)  #n[-10,-9,-8,...,0,1,2,3,...,9] 
shift_time = 4        #n0=4

ramp=Ramp_signal(n,shift_time)  #r[n-4]
plt.plot(n,ramp)
plt.xlabel('t')

plt.ylabel('$r(t)$')
plt.title('Continus Unit Ramp Signal $r(t-a)$')
plt.grid()
plt.show()
```

# Unit Exponential  Signal
$$e^{an}$$ 

### a is real.

```python
def Exponential_signal(n,a):
    exponential = np.exp(n*a)        
            
    return exponential

LL = -10
UL = 10
n = np.arange(LL,UL,1)  #n[-10,-9,-8,...,0,1,2,3,...,9] 
a = 2
exponential = Exponential_signal(n,a)
plt.figure(figsize=(12,8))
plt.stem(n,exponential)
plt.xlabel('n')
plt.xticks(np.arange(LL, UL, 1))

plt.ylabel('$e^{an}$')
plt.title('Unit Ramp Signal $e^{an}$')
plt.show()
```

 # -----------------------------------------------------------------
 $$A\sin(wt+\phi)$$
 $$T = \frac{1}{f}$$
 $$w = \frac{2\pi}{T} = 2 \pi f$$

```python
#sampling rate
sr = 100

#sampling interval 
ts = 1/sr
t = np.arange(0,1,ts)

#signal x_1(t)=sin(wt+q) that q=0 f=5 hz
#w = 2*pi*f
q1 = 0
f1 = 5 #frequence = 5 hz
w1 = 2*np.pi*f1
x1 = np.sin(w1*t+q1)

#signal x_2(t)=sin(wt+q) that q=0 f=10 hz
#w = 2*pi*f
q2 = 0
f2 = 10 #frequence = 5 hz
w2 = 2*np.pi*f2
x2 = np.sin(w2*t+q2)

#ploting

#pot x_1(t)
plt.figure(figsize = (12, 8))
plt.subplot(211)
plt.plot(t, x1, 'b')
plt.ylabel('Amplitude')


plt.subplot(212)
plt.plot(t, x2, 'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')
plt.show()
```
